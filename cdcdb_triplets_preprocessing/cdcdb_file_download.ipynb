```python
import os

# this is the directory where you download the zip file of the latest version of cdcdb triplets
# and also put this notebook in the directory
cdcdb_dir = cdcdb_dir
# check if upzipped already
if not os.path.exists(cdcdb_dir+'dcdb.sqlite'):
    print("Unzipping downloaded cdcdb zip file...")
    # upzip the zip file
    os.system(f"unzip {cdcdb_dir}download_latest_version.zip")
    print("Finishing unzipping cdcdb zip file!")
else:
    print("cdcdb zip file already unzipped.")
```

```python
import sqlite3
import pandas as pd

db_name = 'dcdb.sqlite'
# Connect to the SQLite database
conn = sqlite3.connect(cdcdb_dir+db_name)

# Query to get all table names
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Available tables:", tables)

# Read the table into a pandas DataFrame
table_name = 'web_preview'
DB_trial = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
table_name = 'mesh_terms'  # replace with your table name
mesh_terms = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

# Close the connection
conn.close()

# save the tables
print("")
DB_trial.to_csv(cdcdb_dir+'drug_trial.csv', index=False)
print("save normalized cdcdb drug combinations in drug_trial.csv!")
mesh_terms.to_csv(cdcdb_dir+'mesh_terms.csv', index=False)
print("save nct_id with mesh terms in mesh_terms.csv!")
```
